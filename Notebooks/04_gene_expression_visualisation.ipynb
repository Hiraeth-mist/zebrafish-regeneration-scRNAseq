```python
# To visualize the expression of candidate genes (taf1b, rrn3, sprtn, polr1d) and known marker genes across developmental timepoints during zebrafish fin regeneration using UMAP and other expression plots.
```

```python
# ============================================
# Zebrafish Fin Regeneration - Gene Expression (by clusters)
# ============================================

import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns

# Load your integrated dataset (.h5ad file)
adata = sc.read("merged_regeneration_data_clustered.h5ad")

# List of genes to visualize
genes_of_interest = ["taf1b", "rrn3", "sprtn", "polr1d"]

# Add marker genes (edit below with your relevant markers)
marker_genes = ["msx1b", "wnt5b", "mmp9", "col1a1a"]  # Example zebrafish fin markers

# Combine
gene_list = genes_of_interest + marker_genes

# -----------------------------------------
# 1️⃣ UMAP Plots (Per Gene)
# -----------------------------------------
sc.pl.umap(adata, color=gene_list, ncols=3, cmap="viridis", save="_gene_umaps.png")

# -----------------------------------------
# 2️⃣ Violin Plots (Per Gene across Clusters)
# -----------------------------------------
sc.pl.violin(adata, keys=gene_list, groupby="leiden_clusters_shifted", jitter=0.4, multi_panel=True, save="_gene_violin_cluster.png")
# Or if you have clusters: groupby="leiden" or "clusters"

# -----------------------------------------
# 3️⃣ Dot Plot (Compact Summary View)
# -----------------------------------------
sc.pl.dotplot(adata, var_names=gene_list, groupby="leiden_clusters_shifted", save="_gene_dotplot_cluster.png")

# -----------------------------------------
# 4️⃣ Heatmap (For Subset of Cells)
# -----------------------------------------
sc.pl.heatmap(adata, var_names=gene_list, groupby="leiden_clusters_shifted", cmap="viridis", save="_gene_heatmap_cluster.png")
```
